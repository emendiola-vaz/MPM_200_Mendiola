# Python Packages and Scientific Computing Libraries

"""
Python Packages Overview:
-------------------------
Packages in Python are collections of modules that provide reusable code and functionality.
They allow developers to organize and share complex code libraries that can be easily imported

Key Characteristics of Packages:
- Organized hierarchically
- Contain multiple related modules
- Can be installed using package managers like pip
- Extend Python's core functionality

--------------------------------------
1. Pandas (pd):

   - Provides DataFrame: a 2D labeled data structure with columns of potentially different types

     * Complex data transformations

     * Line plots, scatter plots, bar charts, histograms

     * Support for multiple plot types and graph formats
     * Integration with NumPy for data visualization

Practical Significance:
----------------------
These libraries form the backbone of data science and scientific computing in Python,

## Perform a vertical join of Mammals and Birds data frames:  
* Identify and select only the common columns present in both data frames.  
* Use `pd.concat()` or `df.append()` to create a larger dataframe called **metadata**.  
* Ensure the resulting dataframe has the correct number of rows and columns.  

```python
# Creating working copies of dataframes
birds_speciesonly = birds_speciesonly.copy()
mammals_meta = mammals_meta.copy()
```

```python
# Concatenating Genus and Species into a Scientific Name column in the mammals_meta dataframe
mammals_meta["Scientific Name"] = (
    mammals_meta["Genus"] + " " + mammals_meta["Species"])
```

```python
mammals_meta[["Genus", "Species", "Scientific Name"]].head(20)
```

```python
birds_speciesonly["Scientific Name"] = birds_speciesonly["Scientific name"]
mammals_meta["Scientific Name"] = mammals_meta["Scientific Name"]
```

```python
birds_common = birds_speciesonly[["Scientific Name", "Family", "Order"]].copy()
mammals_common = mammals_meta[["Scientific Name", "Family", "Order"]].copy()
```

```python
metadata = pd.concat([birds_common, mammals_common], axis=0)
metadata.shape
```

## Merge the metadata dataframe with the Flavivirus dataframe:  
* Experiment with different join types: `inner`, `outer`, `left`, and `right`.  
* For each join type, record the resulting dataframe's shape (rows and columns).  
* Determine the **theoretically correct size of the merged dataframe if done manually**.

```python
# INNER
inner_merge = pd.merge(df, metadata, on='Scientific Name', how='inner')
inner_merge.shape
```

```python
# OUTER
outer_merge = pd.merge(df, metadata, on='Scientific Name', how='outer')
outer_merge.shape
```

```python
# RIGHT
right_merge = pd.merge(df, metadata, on='Scientific Name', how='right')
right_merge.shape
```

```python
# LEFT
left_merges = pd.merge(df, metadata, on='Scientific Name', how='left')
left_merges.shape
```

```python
metadata_unique = metadata.drop_duplicates(subset='Scientific Name')
left_merge = pd.merge(df, metadata_unique, on='Scientific Name', how='left')
left_merge.shape
```

# Explanation for Left Merge
A left merge is the most appropriate join type.
The initial code _left_merges_ outputs 881 rows. This is because pandas was creating an additional row for every duplicated species in the metadata. The second code, _left_merges_ (with an 's'), makes sure that duplicates are dropped in order to keep the original 570 rows of df, which should be the theoretically correct number of rows for this left merge.


### Create a pivote table,
